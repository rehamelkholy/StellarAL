```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    'axes.titlesize': 16,     # Title font size
    'axes.labelsize': 16,     # X and Y axis labels font size
    'xtick.labelsize': 14,    # X tick labels font size
    'ytick.labelsize': 14,    # Y tick labels font size
    'legend.fontsize': 16,    # Legend font size
    'figure.titlesize': 16    # Figure title font size
})
```

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))

# Random model line (AUC = 0.50)
plt.plot([0, 1], [0, 1], color='r', lw=2, linestyle='-', label='Random model (AUC = 0.50)')

# Good model line (AUC ~ 0.79) - This is a placeholder
theta = np.linspace(np.pi / 2, np.pi, 100)
good_model_fpr = np.cos(theta) + 1
good_model_tpr = np.sin(theta)
plt.plot(good_model_fpr, good_model_tpr, color='b', lw=2, linestyle='-', label='Good model (AUC ~ 0.79)')

# Perfect model line (AUC = 1.00)
plt.plot([0, 0, 1], [0, 1, 1], color='g', lw=2, linestyle='-', label='Perfect model (AUC = 1.00)')

# Customize the plot
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)

plt.savefig('ROC.pdf');
```
